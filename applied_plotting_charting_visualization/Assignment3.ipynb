In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import sem,t

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df.head()
```

```python

def compute_confidence_interval(data,confidence=0.95):
    '''
        Compute confidence interval on data series according to procedure outlined in:
        https://kite.com/python/examples/702/scipy-compute-a-confidence-interval-from-a-dataset
    '''
    mean = np.average(data)
    error = sem(data)
    n = len(data)
    
    interval = error*t.ppf((1+confidence)/2,n-1)
    
    return interval
```

```python
# Generate data
year = df.index.values
mean = df.T.apply(np.mean)
error = df.T.apply(compute_confidence_interval)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook


```

```python

def make_plot():
    # Generate plot
    plt.bar(year,mean,yerr=error)

    # Clean up plot a bit

    ax = plt.gca()
    _ = ax.set_xticks(year)
    _ = ax.set_xlabel("Year")
    _ = ax.set_ylabel("Count")
    _ = ax.tick_params(axis = 'x',which = 'both',bottom = False,top = False)
```

```python
# Determine color for each bar 
def color_graph(user_input):
    plt.gcf().clf()
    make_plot()
    for avg,err in zip(mean,error):
        if user_input < avg-err:
            colors.append("red")
        elif user_input <= avg+err:
            colors.append("#FFFFFF")
        else:
            colors.append("blue")
    ax = plt.gca()
    # Draw line
    x1,x2 = ax.get_xlim()
    y1,y2 = ax.get_ylim()
    _ = plt.plot((x1, x2), (user_input, user_input), 'k--',linewidth=0.5)
    #_ = plt.text(x1,0.90*y2,"User input: {}".format(user_input))
        
    # Update plot
    plt.bar(year,mean,yerr=error,color=colors)
    
    
# Conditional coloring

user_input = 39000.

colors = []


color_graph(user_input)
```

```python
# Want colormap seismic
import matplotlib.cm as cm

def draw_line(user_input):
    # Draw line
    ax = plt.gca()
    x1,x2 = ax.get_xlim()
    y1,y2 = ax.get_ylim()
    _ = plt.plot((x1, x2), (user_input, user_input), 'k--',linewidth=0.5)
    
def color_graph_gradient(user_input):
    plt.figure()
    make_plot()
    interval_ratio = lambda avg,err,user_input: 1- ((avg+err) - user_input)/(2*err)
    cmap = cm.get_cmap("seismic_r")
    
    # Get colors based upon how much of the 'error envelope' (=mean+/-error) is covered by user input
    colors = interval_ratio(mean,error,user_input)
    colors[colors>1] = 1.0
    colors[colors<0] = 0.0

    # Update plot
    _ = plt.bar(year,mean,yerr=error,color=cmap(colors))

    # Add color bar
    sm = cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(0,1))
    sm.set_array([])
    cbar = plt.colorbar(sm)
    
    draw_line(user_input)

color_graph_gradient(user_input)
```
